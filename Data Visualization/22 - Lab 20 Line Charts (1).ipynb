### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here

comp_age = df.groupby("Age")["ConvertedCompYearly"].median()

ax = comp_age.plot(kind="line", marker="o", figsize=(10,6))
plt.title("Median Yearly Compensation by Age Group")
plt.xlabel("Age")
plt.ylabel("Median Compensation (USD)")
plt.grid(True)

# Rotate x labels vertically
plt.xticks(rotation=90)

plt.show()


```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here

def age_midpoint(age):
    if "-" in str(age):
        lo, hi = age.split("-")
        return (int(lo) + int(hi.split()[0])) / 2
    elif "Under" in str(age):
        return 17
    elif "65" in str(age):
        return 70
    elif "Prefer" in str(age):
        return None
    else:
        return None

df["Age_num"] = df["Age"].apply(age_midpoint)


comp_age_num = df.groupby("Age_num")["ConvertedCompYearly"].median().dropna()


comp_age_25_45 = comp_age_num.loc[(comp_age_num.index >= 25) & (comp_age_num.index <= 45)]


ax = comp_age_25_45.plot(kind="line", marker="o", color="orange", figsize=(10,6))
plt.title("Median Yearly Compensation (Ages 25–45)")
plt.xlabel("Age")
plt.ylabel("Median Compensation (USD)")
plt.grid(True)
plt.xticks(rotation=0)  # Keep horizontal since numbers are short
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here

def exp_to_num(exp):
    if pd.isna(exp): return None
    exp = str(exp)
    if "-" in exp: 
        lo, hi = exp.split("-")
        return (int(lo) + int(hi.split()[0])) / 2
    elif "Less than" in exp:
        return 0.5
    elif "More than" in exp:
        return int(''.join(filter(str.isdigit, exp))) 
    else:
        try:
            return float(exp)
        except:
            return None

df["WorkExp_num"] = df["WorkExp"].apply(exp_to_num)


exp_sat = df.groupby("WorkExp_num")["JobSatPoints_6"].median().dropna()

# Plot
ax = exp_sat.plot(kind="line", marker="o", figsize=(10,6), color="green")
plt.title("Median Job Satisfaction by Experience Level")
plt.xlabel("Years of Experience")
plt.ylabel("Median Job Satisfaction (Q6)")
plt.grid(True)
plt.show()
```
